## Helper functions

```python
def rel_error(x, y):
    """ returns relative error """
    return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))

def eval_numerical_gradient_array(f, x, df, h=1e-5):
    """
    Evaluate a numeric gradient for a function that accepts a numpy
    array and returns a numpy array.
    """
    grad = np.zeros_like(x)
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        ix = it.multi_index

        oldval = x[ix]
        x[ix] = oldval + h
        pos = f(x).copy()
        x[ix] = oldval - h
        neg = f(x).copy()
        x[ix] = oldval

        grad[ix] = np.sum((pos - neg) * df) / (2 * h)
        it.iternext()
    return grad
```

#RNN: step backward (10 points)

```python
def rnn_step_backward(dnext_h, cache):
    """
    Backward pass for a single timestep of an RNN.

    Inputs:
    - dnext_h: Gradient of loss with respect to next hidden state
    - cache: Cache object from the forward pass

    Returns a tuple of:
    - dx: Gradients of input data, of shape (N, D)
    - dprev_h: Gradients of previous hidden state, of shape (N, H)
    - dWx: Gradients of input-to-hidden weights, of shape (D, H)
    - dWh: Gradients of hidden-to-hidden weights, of shape (H, H)
    - db: Gradients of bias vector, of shape (H,)
    """
    dx, dprev_h, dWx, dWh, db = None, None, None, None, None
    ##############################################################################
    # TODO: Implement the backward pass for a single step of a RNN.              #
    #                                                                            #
    # HINT: For the tanh function, you can compute the local derivative in terms #
    # of the output value from tanh.                                             #
    ##############################################################################
    x, prev_h, Wx, Wh, b = cache

    # use of hint to compute derivative of tanh
    next_h = np.tanh(np.dot(x, Wx) + np.dot(prev_h, Wh) + b)
    dtan = 1 - np.square(next_h)

    dx = np.dot(dnext_h * dtan, np.transpose(Wx))       # dx = de/dx = de/dht * dht/dx = dnext_h * Wx.T *(1-next_h^2)
    dprev_h = np.dot(dnext_h * dtan, Wh.T)              # dprev_h = de/dprev_h = de/dht *  dht/dprev_h = dnext_h * Wh.T *(1-next_h^2)
    dWx = np.dot(np.transpose(x),(dnext_h * dtan))      # dWx = de/dWx = de/dht * dht/dWx = dnext_h * (1-next_h^2) * np.transpose(x)
    dWh = np.dot(np.transpose(prev_h),(dnext_h * dtan)) # dWh = de/dWh = de/dht * dht/dWh = dnext_h * (1-next_h^2) * prev_h
    db = np.sum(dnext_h * dtan, axis=0)                 # db = de/db = de/dht * dht/db = dnext_h * (1-next_h^2)

    ##############################################################################
    #                               END OF YOUR CODE                             #
    ##############################################################################
    return dx, dprev_h, dWx, dWh, db
```

# RNN: forward (10 points)
Now that you have implemented the forward and backward passes for a single timestep of an RNN, you will combine these pieces to implement an RNN that process an entire sequence of data.

```python
def rnn_forward(x, h0, Wx, Wh, b):
    """
    Run an RNN forward on an entire sequence of data. We assume an input
    sequence composed of T vectors, each of dimension D. The RNN uses a hidden
    size of H, and we work over a minibatch containing N sequences. After running
    the RNN forward, we return the hidden states for all timesteps.

    Inputs:
    - x: Input data for the entire timeseries, of shape (N, T, D).
    - h0: Initial hidden state, of shape (N, H)
    - Wx: Weight matrix for input-to-hidden connections, of shape (D, H)
    - Wh: Weight matrix for hidden-to-hidden connections, of shape (H, H)
    - b: Biases of shape (H,)

    Returns a tuple of:
    - h: Hidden states for the entire timeseries, of shape (N, T, H).
    - cache: Values needed in the backward pass
    """
    h, cache = None, None
    ##############################################################################
    # TODO: Implement forward pass for a RNN running on a sequence of            #
    # input data. You should use the rnn_step_forward function that you defined  #
    # above. You can use a for loop to help compute the forward pass.            #
    ##############################################################################
    N, T, D = x.shape
    H = h0.shape[1]


    h = np.zeros((N, T, H))
    cache = []

    prev_h = h0
    for t in range(T):
      x_t = x[:,t,:]
      next_h, cache_t = rnn_step_forward(x_t, prev_h, Wx, Wh, b)
      cache.append(cache_t)
      h[:, t, :] = next_h
      prev_h = next_h

    ##############################################################################
    #                               END OF YOUR CODE                             #
    ##############################################################################
    return h, cache
```

# RNN: backward (10 points)

```python
def rnn_backward(dh, cache):
    """
    Compute the backward pass for an RNN over an entire sequence of data.

    Inputs:
    - dh: Upstream gradients of all hidden states, of shape (N, T, H)

    Returns a tuple of:
    - dx: Gradient of inputs, of shape (N, T, D)
    - dh0: Gradient of initial hidden state, of shape (N, H)
    - dWx: Gradient of input-to-hidden weights, of shape (D, H)
    - dWh: Gradient of hidden-to-hidden weights, of shape (H, H)
    - db: Gradient of biases, of shape (H,)
    """
    dx, dh0, dWx, dWh, db = None, None, None, None, None
    ##############################################################################
    # TODO: Implement the backward pass for a RNN running an entire              #
    # sequence of data. You should use the rnn_step_backward function that you   #
    # defined above. You can use a for loop to help compute the backward pass.   #
    ##############################################################################
    N, T, H = dh.shape
    D = cache[0][0].shape[1]

    dx  = np.zeros((N, T, D))
    dh0 = np.zeros((N, H))
    dWx = np.zeros((D, H))
    dWh = np.zeros((H, H))
    db  = np.zeros((H))

    dprev_h_t  = np.zeros((N, H))
    for t in reversed(range(T)):
      cache_t = cache[t]
      dh_t = dh[:,t,:] + dprev_h_t

      dx_t, dprev_h_t, dWx_t, dWh_t, db_t = rnn_step_backward(dh_t, cache_t)
      dx[:,t,:] = dx_t
      dWx += dWx_t
      dWh += dWh_t
      db += db_t

    dh0 = dprev_h_t
    ##############################################################################
    #                               END OF YOUR CODE                             #
    ##############################################################################
    return dx, dh0, dWx, dWh, db
```

##**Question:**
When using an RNN on long sequences, what could happen? What causes this to happen?   **(5 Points)**

**Your answer:**

1. **Difficulty in optimization**: RNNs can be hard to optimize when dealing with long sequences. This is primarily due to the issues of vanishing and exploding gradients. The vanishing gradients problem occurs when the gradients of the loss function approach zero for earlier time steps in the sequence, making it difficult for the model to learn and adjust the weights associated with these time steps during backpropagation. On the other hand, the exploding gradients problem occurs when the gradients become too large, causing the weights in the network to update too drastically and leading to unstable learning and erratic behavior.

2. **Slow training**: Training RNNs on long sequences can be slow. This is because RNNs need to process each step in the sequence one at a time, which can be computationally expensive for long sequences.

3. **Inability to capture long-term dependencies**: Although RNNs are designed to handle sequential data, they often struggle to capture long-term dependencies due to the aforementioned vanishing gradient problem. This means that the model may fail to recognize important relationships between elements that are far apart in the sequence.



##**Question:**
Could this problem be solved by a different model? How does it accomplish this?   **(5 Points)**

**Your answer:**

Yes, the problems associated with RNNs on long sequences can be mitigated by using different models such as **Long Short-Term Memory (LSTM)** networks and **Gated Recurrent Unit (GRU)** networks.

1. **LSTM**: LSTMs are a special kind of RNN that are designed to avoid the long-term dependency problem. They accomplish this through a complex cell structure that includes an input gate, an output gate, and a forget gate. These gates determine how much information should be kept or discarded at each time step, allowing the model to learn what information is important to keep and what can be forgotten, thereby mitigating the vanishing gradient problem.

2. **GRU**: GRUs are a simplified version of LSTMs that combine the forget and input gates into a single "update gate". They also merge the cell state and hidden state, resulting in a simpler and faster model. Like LSTMs, GRUs are designed to better capture long-term dependencies compared to standard RNNs.

Both LSTMs and GRUs have been shown to perform well on tasks involving long sequences, making them a popular choice for many applications in natural language processing, speech recognition, and more.

```python
class ConvNet(nn.Module):
    def __init__(self, embed_size):
        super(ConvNet, self).__init__()
        resnet = models.resnet152() # construct an nn.Sequential model without the last resnet152 layer
        modules = list(resnet.children())[:-1]
        self.resnet = nn.Sequential(*modules)
        # make sure you define each of the None parameters
        self.linear = None
        self.bn = None
        #############################################################################
        # TO DO:                                                                    #
        # create a new sequential model which includes the resnet and               #
        # add a new fully connected layer that outputs a vector with the size of    #
        # the wanted embedding. Next, you should add a batchnorm layer with         #
        # momentum=0.01 (BatchNorm1d parameter).                                    #
        # This function has no return value.                                        #
        #############################################################################

        # Create a fully connected layer
        self.linear = nn.Linear(resnet.fc.in_features, embed_size)

        # Create a BatchNorm layer with momentum
        self.bn = nn.BatchNorm1d(embed_size, momentum=0.01)

        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################

    def forward(self, images):
        #############################################################################
        # TO DO:                                                                    #
        # Define the forward propagation. You need to pass an image through the     #
        # network and extract the feature vector. In this case, when using a        #
        # perdefined network, you don't want to change it's weights.                #
        # The rest of the layers you defined should accepts gradients for them to   #
        # improve during training. Make sure you are inputing a correct shape       #
        # to the batchnorm layer.                                                   #
        # This function returns the features of the image.                          #
        #############################################################################

        features = self.resnet(images)

        # Flatten features for classification
        x = features.view(features.shape[0], -1)

        x = self.linear(x)
        x = self.bn(x)

        return x

        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################


class LSTM(nn.Module):
    def __init__(self, embed_size, hidden_size, vocab_size, num_layers, max_seq_length=20):
        super(LSTM, self).__init__()
        self.max_seg_length = max_seq_length
        self.embed = None
        self.lstm = None
        self.linear = None
        #############################################################################
        # TO DO:                                                                    #
        # Define the hyper-parameters and the layers of the pretrained LSTM.        #
        # Create an Embedding layer that accepts the output of the                  #
        # feature extractor.  Next, the built-in LSTM architecture in PyTorch.nn    #
        # with the proper inputs (use the built-in documentation tool in Jupyter    #
        # or just look at the official documentation online).                       #
        # Define an additional linear layer that comes after the LSTM and outputs   #
        # a vector that will support the size of our vocabulary.                    #
        # This function has no return value.                                        #
        #############################################################################

        # Create an Embedding layer
        # This layer will be used to convert word indices into dense vectors of fixed size (embeddings).
        # vocab_size: the number of unique words in the vocabulary
        # embed_size: the dimensionality of the embedding space
        self.embed = nn.Embedding(vocab_size, embed_size)

        # Create an LSTM layer
        # This layer will process sequences of word embeddings and learn to capture temporal dependencies.
        # embed_size: the size of the input features (word embeddings)
        # hidden_size: the number of features in the hidden state of the LSTM
        # num_layers: the number of recurrent layers in the LSTM
        # batch_first=True: input and output tensors are provided as (batch_size, seq_length, feature_dim)
        self.lstm = nn.LSTM(embed_size, hidden_size, num_layers, batch_first=True)

        # Create a fully connected layer
        # This layer will take the output of the LSTM and produce the final output predictions.
        # hidden_size: the size of the input features (output of the LSTM)
        # vocab_size: the size of the vocabulary, i.e., the number of possible output words
        self.linear = nn.Linear(hidden_size, vocab_size)

        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################

    def sample(self, features, states=None):
        sampled_ids = [] # Initialize the list that will contain the predicted word IDs
        inputs = features.unsqueeze(1) # Unsqueeze features for sequence length; 'inputs' shape: (batch_size, 1, feature_size)
        #############################################################################
        # TO DO:                                                                    #
        # Generate captions for a given image features.                             #
        # First, obtain the output of the LSTM network.                             #
        # Next, use the hidden states to obtain the most probable word and store    #
        # all the word predictions in the sampled_ids list. Don't forget to update  #
        # the inputs for each timestep to continue making predictions based on the  #
        # words you are alreaedy predicted.                                         #
        # Make sure you keep track of the dimensions of the inputs and outputs,     #
        # since PyTorch expects tensors with a batch dimension. You can use the     #
        # methods .squeeze() and .unsqueeze()                                       #
        # This function returns the list of predicted words.                        #
        #############################################################################

        # Loop through each time step
        for _ in range(self.max_seg_length):
            # Pass the inputs through the LSTM layer
            lstm_out, states = self.lstm(inputs, states)

            # Pass the LSTM output through the linear layer
            outputs = self.linear(lstm_out.squeeze(1))

            # Get the index of the maximum value as the predicted word
            _, predicted = outputs.max(1)

            # Append the predicted word ID to the sampled_ids list
            sampled_ids.append(predicted.item())

            # Update the inputs for the next time step
            inputs = self.embed(predicted).unsqueeze(1)

        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################
        return sampled_ids
```

```python
def load_image(image_path, transform=None):
    image = Image.open(image_path)
    image = image.resize([224, 224], Image.Resampling.LANCZOS)

    if transform is not None:
        image = transform(image).unsqueeze(0)

    return image

class Vocabulary(object):
    """Simple vocabulary wrapper."""
    def __init__(self):
        self.word2idx = {}
        self.idx2word = {}
        self.idx = 0

    def add_word(self, word):
        if not word in self.word2idx:
            self.word2idx[word] = self.idx
            self.idx2word[self.idx] = word
            self.idx += 1

    def __call__(self, word):
        if not word in self.word2idx:
            return self.word2idx['<unk>']
        return self.word2idx[word]

    def __len__(self):
        return len(self.word2idx)
```

There are several scenarios where the image captioning model might fail:

1. **Complex Images**: If an image has many objects, intricate details, or complex scenes, the model might not be able to accurately identify all elements and their interactions, leading to incomplete or incorrect captions.

2. **Unseen Objects or Contexts**: The model is trained on a specific dataset. If it encounters objects or contexts in the test images that it has not seen during training, it might fail to recognize them.

3. **Ambiguity**: If an image is ambiguous or can be interpreted in multiple ways, the model might not choose the interpretation a human would choose. For example, an image of a person standing next to a tree could be captioned as "a person standing next to a tree" or "a person in a park". Both are technically correct, but one might be more appropriate given the context.

4. **Poor Image Quality**: If an image has low resolution, is blurred, or has poor lighting, the model might fail to correctly identify the objects in the image, leading to incorrect captions.

5. **Abstract Concepts**: The model might struggle with abstract concepts, emotions, or symbolic imagery that humans can understand. For example, a picture of a person with a tear rolling down their cheek could symbolize sadness, but the model might simply describe it as "a person with a tear on their face".

These failures occur because while the model can extract features from images and generate captions based on those features, it does not truly "understand" the image in the way humans do. It lacks the ability to reason abstractly, to use common sense, or to draw on a wide range of world knowledge that humans naturally use when interpreting images. Furthermore, the model's performance is heavily dependent on the quality and diversity of the data it was trained on.
